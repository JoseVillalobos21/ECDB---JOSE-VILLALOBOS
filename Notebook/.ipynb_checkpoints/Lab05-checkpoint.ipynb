```python
# Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga del DataSet
df = pd.read_csv("netflix_titles.csv")
```

```python
# Identificar valores nulos (totales por columna)
df.isnull().sum()
```

```python
# Calcular el porcentaje de valores faltantes por columna
(df.isnull().sum() / len(df)) * 100
```

```python
# Detectar registros duplicados
df.duplicated().sum()
```

```python
# --- ESTRATEGIA DE LIMPIEZA COMPLETA ---

# 1. Corrección del desfase en la columna 'duration' (filas donde duration es nula y el valor está en 'rating')
df.loc[df['duration'].isnull(), 'duration'] = df['rating']
# Reemplazamos el valor desfasado en 'rating' por 'NR' (Not Rated)
df.loc[df['rating'].str.contains('min', na=False), 'rating'] = 'NR'

# 2. Imputación de nulos en columnas categóricas con valores faltantes menores al 1%
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])

# 3. Imputación de nulos en columnas con alto porcentaje de faltantes (director, cast, country)
df['director'] = df['director'].fillna('Unknown Director')
df['cast'] = df['cast'].fillna('Unknown Cast')
df['country'] = df['country'].fillna('Unknown Country')

# Verificar que ya no hay valores nulos
df.isnull().sum()
```

#### Justificación de decisiones de limpieza:
1. **Desfase de Duración:** Se detectó que tres registros (especiales de comedia de Louis C.K.) tenían la duración en la columna `rating`. Se transfirió el valor y se redefinió la clasificación como `NR` (Not Rated).
2. **Imputación por Moda:** Para `rating` y `date_added` los valores nulos representan menos del 0.2%, por lo que imputar con la moda es seguro y no sesga los datos.
3. **Uso de etiquetas 'Unknown':** Para `director`, `cast` y `country` el volumen de nulos es alto (~30% en director). Eliminar estas filas implicaría perder valiosa información descriptiva de otras columnas, por lo que se decidió transformarlos asignando un valor centinela (`Unknown`).

```python
# Gráfica 1: Distribución del tipo de contenido (Películas vs Series de TV)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type', hue='type', palette='Set2', legend=False)
plt.title('Distribución de Contenido en Netflix (Movie vs TV Show)')
plt.xlabel('Tipo de Contenido')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

```python
# Gráfica 2: Top 10 países productores de contenido (usando el país principal)
df['primary_country'] = df['country'].str.split(',').str[0]
top_countries = df['primary_country'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, hue=top_countries.index, palette='viridis', legend=False)
plt.title('Top 10 Países Productores de Contenido en Netflix')
plt.xlabel('Cantidad de Títulos')
plt.ylabel('País')
plt.show()
```

```python
# Gráfica 3: Distribución del año de lanzamiento (release_year) a partir del año 2000
plt.figure(figsize=(10, 5))
sns.histplot(data=df[df['release_year'] >= 2000], x='release_year', kde=True, bins=22, color='crimson')
plt.title('Evolución de Años de Lanzamiento (2000 - Presente)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

```python
# Gráfica 4: Top 10 géneros más populares en la plataforma
genres = df['listed_in'].str.split(', ').explode().value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=genres.values, y=genres.index, hue=genres.index, palette='coolwarm', legend=False)
plt.title('Top 10 Géneros con Mayor Presencia en Netflix')
plt.xlabel('Cantidad de Títulos')
plt.ylabel('Género')
plt.show()
```

### ---- 4. Interpretación de patrones (CAPA DE CONOCIMIENTO)

```python
# Conclusión 1: Proporción del catálogo dedicada a películas vs. series
proporcion_tipo = df['type'].value_counts(normalize=True) * 100
print("Porcentaje por tipo de contenido:")
print(proporcion_tipo)
```

```python
# Conclusión 2: Centralización de la producción en el mercado estadounidense
pct_eeuu = (df['primary_country'] == 'United States').mean() * 100
print(f"Porcentaje de títulos producidos principalmente en EE.UU.: {pct_eeuu:.2f}%")
```

```python
# Conclusión 3: Concentración temporal de la oferta de contenido disponible
pct_reciente = (df['release_year'] >= 2015).mean() * 100
print(f"Porcentaje de títulos con año de lanzamiento >= 2015: {pct_reciente:.2f}%")
```

#### Conclusiones Derivadas del Análisis (Capa de Conocimiento):
1. **Dominio del Formato Largometraje (Películas):** Del análisis se desprende que aproximadamente el **69.6%** del catálogo de Netflix está compuesto por películas (`Movie`), mientras que el **30.4%** son series (`TV Show`). Esto indica que el modelo de negocio de la plataforma sigue estando fuertemente anclado en películas individuales a pesar del crecimiento en popularidad de las series.
2. **Liderazgo Hegemónico de Estados Unidos:** Estados Unidos es, por un margen extremadamente amplio, el principal productor de contenido de Netflix, abarcando más del **30%** de toda la producción del catálogo mundial. Le siguen a gran distancia otros centros de producción como India y el Reino Unido.
3. **Catálogo de Estreno Reciente:** La inmensa mayoría del catálogo disponible es de producción muy reciente. Más del **70%** de los títulos fueron lanzados a partir del año 2015. Esto revela una estrategia enfocada en la frescura y actualización constante del catálogo, desechando o reduciendo el licenciamiento de producciones más antiguas.
