```python
# Importamos la librería pandas para la manipulación de datos
import pandas as pd
```

```python
# Cargamos el dataset en un DataFrame usando Pandas
df = pd.read_csv("../DataSet/test.csv")
```

```python
df
```

```python
# Obtenemos un resumen de las columnas, tipos de datos y valores no nulos
df.info()
```

```python
# Visualizamos las primeras filas para inspeccionar la estructura de los datos
df.head(3)
```

```python
# Calculamos estadísticas descriptivas básicas (media, min, max, etc.)
df.describe(include="all")
```

```python
# Extraemos los nombres de las columnas para revisarlas
df.columns=df.columns.str.strip().str.lower()
```

```python
# Extraemos los nombres de las columnas para revisarlas
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["cabin"], inplace=True)
```

```python
# Extraemos los nombres de las columnas para revisarlas
df.columns
```

```python
df["age"]= df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos el Fare con el valor mas repetido (moda)
```

```python
df["fare"]=df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(6,3))#Tamaño
sns.countplot(df,x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.show()
```

```python
#Graficamos las clases por genero
plt.figure(figsize=(6,3))#Tamaño
sns.countplot(df,x="sex",hue="pclass")
plt.title("Total de Genero por Clases")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por genero y clase
89plt.figure(figsize=(6,4))#Tamaño
sns.violinplot(df,x="sex",y="age", hue="pclass")
plt.title("Rango de edad por genero y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMIENTO
#Los pasajeros de la primera clase pagaban tarifas mas altas
df.groupby("pclass")["fare"].mean()
```
